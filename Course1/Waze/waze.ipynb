## Dataset Overview

This project uses a dataset called **waze_dataset.csv**.  
It contains synthetic data created for this project in partnership with Waze.

- **14,999 rows** – Each row represents one unique user.  
- **13 columns** are included in the dataset.

---

## Column Descriptions

| Column Name | Type | Description |
|-------------|------|-------------|
| **ID** | int | Sequential numbered index |
| **label** | obj | Binary target variable (“retained” vs “churned”) indicating whether a user churned during the month |
| **sessions** | int | Number of times the user opened the app during the month |
| **drives** | int | Number of times the user drove at least 1 km during the month |
| **device** | obj | Type of device used to start a session |
| **total_sessions** | float | Model estimate of total sessions since onboarding |
| **n_days_after_onboarding** | int | Number of days since the user signed up |
| **total_navigations_fav1** | int | Total navigations since onboarding to favorite place 1 |
| **total_navigations_fav2** | int | Total navigations since onboarding to favorite place 2 |
| **driven_km_drives** | float | Total kilometers driven during the month |
| **duration_minutes_drives** | float | Total minutes driven during the month |
| **activity_days** | int | Number of days the user opened the app during the month |
| **driving_days** | int | Number of days the user drove at least 1 km during the month |
